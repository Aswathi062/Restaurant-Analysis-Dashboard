```python
df=pd.read_csv('consumers.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Smoker'].fillna('Yes',inplace=True)
df['Transportation_Method'].fillna('Public Transportation',inplace=True)
df['Marital_Status'].fillna('Married',inplace=True)
df['Occupation'].fillna('Student',inplace=True)
df['Budget'].fillna('Medium',inplace=True)
df['Children'].fillna('Independent',inplace=True)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.to_csv('cleaned_consumers_data.csv',index=False)
```
